```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(os.getcwd())
print(os.listdir())

a = np.loadtxt("sismique.txt")

a_v = a*10**-6
dt = 2

N = len(a_v)
t = np.linspace(0,N*dt,N)

plt.plot(t,a_v)
plt.title("Trace Sismique", color="red")
plt.xlabel("Temps [ms]")
plt.ylabel("Amplitude [V]")
plt.grid(True)
plt.show()
```
