**2.2 Analyse des densités**

```python
# Renommer les colonnes
df = df.rename(columns={
    "Rock type/ VISUAL": "RockType",
    "Density/(gm/cc)": "Density"})
```

```python
df.head()
```

```python
df_density = df[["RockType", "Density"]] #Afficher que le type de roche et la densité
df_density.head()
```

```python
mean_density = df_density.groupby("RockType")["Density"].mean() #Moyenne de densité pour chaque type de roche
print(mean_density)
```

```python
std_density = df_density.groupby("RockType")["Density"].std() #Ecart-type des densités par type de roches
print(std_density)
```

```python
stats_density = df_density.pivot_table(   #Permet de tout synthétiser dans un même tableau
    index="RockType",
    values="Density",
    aggfunc=["min", "max", "mean", "std"])

print(stats_density)
```

```python
plt.figure(figsize=(5,3)) #Existe figsize=(x,y) à inserer dans le plt.figure(...) afin de modifier la hauteur et la largeur de la figure créé. 
df_density["Density"].plot(kind="hist", bins=20, edgecolor="black") #kind permet de définir le type d'histogramme -> voir markdown ci après ; bins pour le nombre de barres ; 
                                                                    #edgecolor pour la couleur de contour des barres. Existe aussi color='color' pour modifier la couleur de remplissage.
plt.title("Distribution des densités (toutes roches confondues)")
plt.xlabel("Densité (g/cm³)")
plt.ylabel("Nombre d'échantillons")
plt.show()
#Ici on fait la distribution de toutes les densités toutes roches confondues, plus intéressant de faire par type de roche : voir ci-dessous
```

La fonction df.plot permet de créer un graphique avec nos données. Le paramètre kind sert à demander le type de graphique. 
hist = histogramme ;
line = courbe (par défaut si on met rien) ;
bar = diagramme en barre ;
scatter = nuage de points ;
pie = diagramme circulaire ;

```python
for rock, group in df_density.groupby("RockType"): # Même corps que tout à l'heure avec boucle for pour intégrer chacun des types de roches
    group["Density"].plot(kind="hist", bins=15, alpha=0.5, label=rock, edgecolor="black") # alpha important ici, permet de mettre en transparence le remplissage
                                                                                          # pour permettre une bonne lecture malgré la superposition de certaines barres

plt.title("Histogramme des densités par type de roche")
plt.xlabel("Densité (g/cm³)")
plt.ylabel("Fréquence")
plt.legend()
plt.show()
```

**2.3 Vitesses en fonction des densités**

```python
df = pd.read_excel("rock_properties_TP4.xlsx", header=2) # Réouverture du fichier toujours sans l'entête pour prendre en plus les vitesses

# Renommer les colonnes pour simplifier le code
df = df.rename(columns={"Rock type/ VISUAL": "RockType","Density/(gm/cc)": "Density",})
```

```python
cols_vitesse = [c for c in df.columns if "V (km/s)" in c and "@200" not in c] + ["V (km/s) @200Mpa"]
print(cols_vitesse)

# Créer une figure avec autant de sous-graphiques qu'il y a de vitesses
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(16, 8))  # 10 graphes (2 lignes × 5 colonnes)
axes = axes.flatten()  # simplifie l’accès aux axes

for i, col in enumerate(cols_vitesse):
    df.plot(kind="scatter",x="Density",y=col,alpha=0.6,ax=axes[i],title=col)
    axes[i].set_xlabel("Densité (g/cm³)")
    axes[i].set_ylabel("Vitesse (km/s)")

plt.suptitle("Vitesse des ondes en fonction de la densité selon la pression", fontsize=14)
plt.tight_layout()
plt.show()
```
